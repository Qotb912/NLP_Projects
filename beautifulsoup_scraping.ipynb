```python
bookUrl = 'http://books.toscrape.com/catalogue/category/books/mystery_3/index.html'
bookRes = req.get(bookUrl)
print(bookRes.text)
```

```python
bookSoup = BeautifulSoup(bookRes.text,'html.parser')
# booksData =  bookSoup.find_all('li',attrs={'class': 'col-xs-6 col-sm-4 col-md-3 col-lg-3'})
articles =  bookSoup.find_all('article')
articles
```

```python
rating_dic = {'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
```

```python
from os import name
import pandas as pd
booksdf = pd.DataFrame(None,columns=['title','price','rating'])
booksdf
```

```python
title = articles[0].find('h3').find('a').get('title')
price = float(articles[0].find('p',attrs ={'class':'price_color'}).get_text()[2:])
# rating = articles[0].find('p',attrs={'class':'star-rating'}).attrs['class'][1]
rating = rating_dic[articles[0].find('p',attrs={'class':'star-rating'}).get('class')[-1]]
# booksdf.combine(pd.DataFrame({'title':title,'price':price,'rating':rating} ),index=None)
booksdf = booksdf.append( {'title':title,'price':price,'rating':rating} ,ignore_index=True)
booksdf.head()
```

```python
for article in articles:
  title = articles[0].find('h3').find('a').get('title')
  price = float(articles[0].find('p',attrs ={'class':'price_color'}).get_text()[2:])
  rating = rating_dic[articles[0].find('p',attrs={'class':'star-rating'}).get('class')[-1]]
  booksdf = booksdf.append( {'title':title,'price':price,'rating':rating} ,ignore_index=True)
  
```
